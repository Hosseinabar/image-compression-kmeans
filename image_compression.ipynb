```python
import numpy as np
import copy
import matplotlib.pyplot as plt
from matplotlib import image
from Kmeans import Kmeans
```

```python
def reduce_img(image, prec):
    
    m = image.shape[0]
    n = image.shape[1]
    l = image.shape[2]
    
    rimg = image.reshape(m * n, l)
    number_of_iterations = 20
    k = prec

    centers, labeled = Kmeans(k, rimg, number_of_iterations)
    for i, label in enumerate(labeled):
        rimg[i] = centers[label]

    return rimg.reshape(m, n, l)
```

```python
img_path = 'data/sample_img1.png'
img = image.imread(img_path)
plt.imshow(img)
plt.show()
```

```python
k = 2
reduced_img = reduce_img(img, k)
plt.imshow(reduced_img)
rimg_path = "data/reduced_image_k2.png"
plt.imsave(rimg_path, reduced_img)
```
